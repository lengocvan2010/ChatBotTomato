```python
import fitz
import pytesseract
from PIL import Image
import io
import os
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import re

pytesseract.pytesseract.tesseract_cmd = '/usr/bin/tesseract'

count = 0
def extract_text_from_pdf(pdf_path):
    global count
    count=count+1
    document = fitz.open(pdf_path)
    text = ""

    for page_num in range(len(document)):
        page = document.load_page(page_num)
        for img_index, img in enumerate(page.get_images(full=True)):
            xref = img[0]
            base_image = document.extract_image(xref)
            image_bytes = base_image["image"]
            image = Image.open(io.BytesIO(image_bytes))
            tmpText = pytesseract.image_to_string(image, lang='vie')
            print(pdf_path+"FILE: "+str(count)+" Page "+str(page_num+1)+"/"+str(len(document)))

            text += tmpText
            #print(tmpText)
    return text

def clean_text(text):
    text = re.sub(r'[^\w\s,!?]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

pdf_folder_path = '/content/drive/MyDrive/Colab Notebooks/data1'

all_texts = ""
for filename in os.listdir(pdf_folder_path):
    if filename.endswith('.pdf'):
        file_path = os.path.join(pdf_folder_path, filename)
        all_texts += extract_text_from_pdf(file_path) + "\n"

cleaned_text = clean_text(all_texts)
```
